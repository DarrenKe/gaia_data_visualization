```python
#import the necessary library
import h5py
import numpy as np
import matplotlib.pyplot as plt
```

```python
#create the original dataset 
data = {}
with h5py.File("gaia_validp.h5") as f:
    for k in f:
        data[k] = f[k][:]
```

```python
#visualize 'ra' vs 'dec' using hexbin function
x=data['ra']
y=data['dec']
xmin=min(x)
xmax=max(x)
ymin=min(y)
ymax=max(y)
# width_y=(ymax-ymin)/2
# y=y/width_y
# ymin=min(y)
# ymax=max(y)


plt.subplot(121)
# plt.hexbin(x, y, cmap=plt.cm.YlOrRd_r)
plt.hexbin(x, y,gridsize=50, cmap='inferno')
plt.axis([xmin, xmax, ymin, ymax])
plt.title("Hexagon binning")
cb = plt.colorbar()
cb.set_label('counts')
```

```python
x=data['phot_g_mean_mag']
y=data['pmra']
xmin=min(x)
xmax=max(x)
ymin=min(y)
ymax=max(y)
# width_y=(ymax-ymin)/2
# y=y/width_y
# ymin=min(y)
# ymax=max(y)


plt.subplot(121)
# plt.hexbin(x, y, cmap=plt.cm.YlOrRd_r)
plt.hexbin(x, y,gridsize=150, cmap='inferno')
plt.axis([xmin, xmax, ymin, ymax])
plt.title("Hexagon binning")
cb = plt.colorbar()
cb.set_label('counts')
```
